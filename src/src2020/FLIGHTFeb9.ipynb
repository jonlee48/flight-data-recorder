Analysis of stall data.

```python
import pandas as pd
import matplotlib as plt
```

```python
input_file_path = r'/home/jumpingjon/fdr/dbf2020/FLIGHT17.TXT' 

#this reads in the data
df_flight_data = pd.read_csv(input_file_path, sep=" ", header=None)

#this adds the columns as a header
df_flight_data.columns = [
    'time', 
    'heading', 
    'pitch', 
    'roll', 
    'xaccel', 
    'yaccel', 
    'zaccel', 
    'temp', 
    'pressure', 
    'altitude', 
    'pedo', 
    'gps']
```

```python
time_pitch_altitude = df_flight_data[['time','altitude','pitch', 'roll']]
df_flight_data = df_flight_data[(df_flight_data.index > 8600) & (df_flight_data.index <9000)]
```

```python
df_flight_data.to_csv(r'/home/jumpingjon/fdr/dbf2020/FLIGHT17.csv', index=False)
```

```python
time_pitch_altitude['pitch'] = time_pitch_altitude['pitch'] - 12
time_pitch_altitude['altitude'] = time_pitch_altitude['altitude'] +2
```
